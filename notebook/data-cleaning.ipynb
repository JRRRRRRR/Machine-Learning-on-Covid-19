# Check NA Value

```python
df.isna().sum() / df.shape[0]
```

```python
# check columns
na_dict = {}
for i in df.columns:
    na_dict[i] = df[i].isna().sum() / df.shape[0]
```

```python
sort_na = sorted(na_dict.items(), key=lambda x: x[1], reverse=True)
```

```python
sort_na
```

```python
# Drop columns with over 30% na values
drop_columns = []
for i in df.columns:
    if df[i].isna().sum()/df.shape[0] > 0.3:
        drop_columns.append(i)
```

```python
df.drop(columns=drop_columns, inplace=True)
```

```python
# check data
df.shape
```

```python
# drop the remaining na values by row
df.dropna(inplace=True)
df.shape
```

# Columns Describe

The whole definition link: https://covidtracking.com/about-data/data-definitions

* dataQualityGrade: To help members of the public understand how well each state is performing, we have assigned each state a data-quality grade based on our assessment of the completeness of their reporting. Our original state grades were based on the presence of the most basic categories of data, and failed to capture our current understanding of “completeness,” so we have updated our grading system accordingly.
* Positive: Total number of confirmed plus probable cases of COVID-19 reported by the state or territory, ideally per the August 5, 2020 CSTE case definition.
* PositiveCasesViral: Total number of unique people with a positive PCR or other approved nucleic acid amplification test (NAAT), as reported by the state or territory.
* hospitalizedIncrease: Daily increase in hospitalizedCumulative, calculated from the previous day’s value.
* deathIncrease: Daily increase in death, calculated from the previous day’s value.

```python
# drop some '0' columns
df.drop(columns=['positiveScore', 'totalTestEncountersViralIncrease'], inplace=True)
```

```python
# drop some rows with erroneous values 
columns = ['deathIncrease', 'hospitalizedIncrease', 'negativeIncrease', 'negativeIncrease', 'positiveIncrease', 
           'totalTestResultsIncrease', 'totalTestsPeopleViralIncrease', 'totalTestsViralIncrease']
for i in columns:
    a = df.loc[df[i] < 0]
    a.reset_index(inplace=True)
    drop_rows = a['index']
    df.drop(index=drop_rows, inplace=True)
```

```python
df.describe()
```

```python
# drop some less informative columns
df.drop(columns=['date', 'dataQualityGrade'], inplace=True)
```
